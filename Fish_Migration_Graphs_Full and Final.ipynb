```python
df = df.drop(columns=['year','month','day'])
```

```python
import matplotlib.pyplot as plt
species_location_counts = df.groupby(['latitude', 'longitude', 'accepted_name'])['num'].sum().reset_index()
species = species_location_counts['accepted_name'].unique()

# Loop through each fish species and plot its count against latitude and longitude
for fish_species in species:
    # Filter the data for the current fish species
    species_data = species_location_counts[species_location_counts['accepted_name'] == fish_species]

    # Create a scatter plot for the current fish species
    plt.figure(figsize=(10, 6))
    scatter = plt.scatter(
        species_data['longitude'],
        species_data['latitude'],
        s=species_data['num'] * 10,  # Size of the points (scaled)
        c=species_data['num'],  # Color of the points based on the number of fish
        cmap='viridis',  # Colormap for colors
        alpha=0.6,  # Transparency of the points
        edgecolor='k'  # Edge color of the points
    )

    # Adding color bar
    plt.colorbar(scatter, label='Number of Fish')

    # Adding title and labels
    plt.title(f'Fish Count of {fish_species} by Location')
    plt.xlabel('Longitude')
    plt.ylabel('Latitude')
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
# Group the data by date and species to get the sum of fish count per species per day
species_daily_counts = df.groupby(['date', 'accepted_name'])['num'].sum().unstack(fill_value=0)

# Sort by date (important for time series)
species_daily_counts = species_daily_counts.sort_index()
```

```python
from statsmodels.tsa.arima.model import ARIMA
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

def predict_coordinates_for_date(df, fish_species, target_date, max_forecast_days=365, debug=True):
    try:
       # Convert target_date to datetime if it's a string
        if isinstance(target_date, str):
            target_date = pd.to_datetime(target_date, errors='coerce')  # <- coerce invalid strings to NaT

        # Check if target_date is still valid
        if pd.isna(target_date):
            print("Error: Invalid target date provided.")
            return None
        species_data = df[df['accepted_name'] == fish_species].copy()
        if len(species_data) == 0:
            print(f"No data found for species: {fish_species}")
            return None
        print(species_data['date'])

        species_data['date'] = pd.to_datetime(species_data['date'])
        last_known_date = species_data['date'].max()
        days_to_forecast = (target_date - last_known_date).days
        
        if days_to_forecast < 0:
            print(f"Target date {target_date.date()} is in the past. Last known date: {last_known_date.date()}")
            return None
        print(f"Target date {target_date.date()} is in the past. Last known date: {last_known_date.date()}")
        
        if days_to_forecast > max_forecast_days:
            print(f"Warning: Requested prediction is {days_to_forecast} days in the future.")
            print(f"Predictions beyond {max_forecast_days} days may be unreliable.")
        daily_lat = species_data.set_index('date')['latitude'].resample('D').mean()
        daily_lon = species_data.set_index('date')['longitude'].resample('D').mean()
        print("hello")
        
        valid_idx = daily_lat.notna() & daily_lon.notna()
        daily_lat = daily_lat[valid_idx]
        daily_lon = daily_lon[valid_idx]
        
        forecast_steps = min(max_forecast_days, ((days_to_forecast // 7) + 1) * 7)
        
        if debug:
            print(f"\nPredicting coordinates for {target_date.date()}")
            print(f"Days to forecast: {days_to_forecast}")
            print(f"Using {forecast_steps} forecast steps")
            print(f"\nLast known position:")
            print(f"Date: {last_known_date.date()}")
            print(f"Latitude: {daily_lat.iloc[-1]:.4f}")
            print(f"Longitude: {daily_lon.iloc[-1]:.4f}")
        
        lat_model = ARIMA(daily_lat, order=(2, 0, 2))
        lon_model = ARIMA(daily_lon, order=(2, 0, 2))
        
        lat_fit = lat_model.fit()
        lon_fit = lon_model.fit()
        
        lat_forecast = lat_fit.forecast(steps=forecast_steps)
        lon_forecast = lon_fit.forecast(steps=forecast_steps)
        
        pred_dates = pd.date_range(start=last_known_date + timedelta(days=1),
                                   periods=forecast_steps,
                                   freq='D')

        predictions_df = pd.DataFrame({
            'date': pred_dates,
            'latitude': lat_forecast,
            'longitude': lon_forecast
        }).set_index('date')
        # Debug information added here
        if debug:
            print(f"target_date type: {type(target_date)} | value: {target_date}")
            print(f"predictions_df head:\n{predictions_df.head()}")
            print(f"predictions_df index type: {type(predictions_df.index)}")
            print(f"Is target_date in predictions_df index? {target_date in predictions_df.index}")
        
        if predictions_df.empty or predictions_df.index.isna().any():
            print("Prediction DataFrame is empty or has NaT in index.")
            return None
        
        predictions_df = predictions_df.reindex(
            predictions_df.index.union([target_date])
        ).sort_index()
        
        predictions_df = predictions_df.interpolate(method='time')
        
        if target_date not in predictions_df.index or pd.isna(predictions_df.loc[target_date, 'latitude']):
            print(f"Could not generate prediction for target date: {target_date}")
            return None
        
        target_pred = predictions_df.loc[target_date]
        
        confidence_metrics = {
            'days_forecasted': days_to_forecast,
            'historical_lat_std': daily_lat.std(),
            'historical_lon_std': daily_lon.std(),
            'historical_lat_range': (daily_lat.min(), daily_lat.max()),
            'historical_lon_range': (daily_lon.min(), daily_lon.max()),
            'prediction_within_bounds': True
        }
        
        if (target_pred['latitude'] < confidence_metrics['historical_lat_range'][0] or
            target_pred['latitude'] > confidence_metrics['historical_lat_range'][1] or
            target_pred['longitude'] < confidence_metrics['historical_lon_range'][0] or
            target_pred['longitude'] > confidence_metrics['historical_lon_range'][1]):
            confidence_metrics['prediction_within_bounds'] = False
        
        confidence_score = max(0, 1 - (days_to_forecast / max_forecast_days))
        confidence_metrics['confidence_score'] = confidence_score
        
        if debug:
            print(f"\nPredicted position for {target_date.date()}:")
            print(f"Latitude: {target_pred['latitude']:.4f}")
            print(f"Longitude: {target_pred['longitude']:.4f}")
            print(f"Confidence score: {confidence_score:.2f}")
            
            plt.figure(figsize=(12, 8))
            plt.scatter(daily_lon, daily_lat, c='blue', alpha=0.1, s=1, label='Historical')
            plt.scatter(daily_lon.iloc[-1], daily_lat.iloc[-1], 
                        c='green', s=100, label=f'Last Known ({daily_lon.iloc[-1]:.4f}, {daily_lat.iloc[-1]:.4f})')
            plt.scatter(target_pred['longitude'], target_pred['latitude'], 
            c='red', s=100, 
            label=f"Prediction ({target_pred['longitude']:.4f}, {target_pred['latitude']:.4f})")
            plt.arrow(daily_lon.iloc[-1], daily_lat.iloc[-1],
                      target_pred['longitude'] - daily_lon.iloc[-1],
                      target_pred['latitude'] - daily_lat.iloc[-1],
                      color='black', width=0.01, head_width=0.05,
                      length_includes_head=True, alpha=0.5)
            plt.title(f'Fish Location Prediction for {target_date.date()}\n{fish_species}')
            plt.xlabel('Longitude')
            plt.ylabel('Latitude')
            plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
            plt.grid(True, alpha=0.3)
            plt.tight_layout()
            plt.show()
        
        return {
            'date': target_date,
            'latitude': target_pred['latitude'],
            'longitude': target_pred['longitude'],
            'confidence_metrics': confidence_metrics
        }
        
    except Exception as e:
        print(f"Error in prediction: {str(e)}")
        return None

# Example usage
if __name__ == "__main__":
    target_date = "2025-05-15"
    
    result = predict_coordinates_for_date(
        df,  # Replace this with your actual DataFrame
        'Trachurus trachurus',
        target_date,
        max_forecast_days=365,
        debug=True
    )
    
    if result:
        print("\nFinal Prediction Summary:")
        print(f"Date: {result['date'].date()}")
        print(f"Latitude: {result['latitude']:.4f}")
        print(f"Longitude: {result['longitude']:.4f}")
        print(f"Confidence Score: {result['confidence_metrics']['confidence_score']:.2f}")
```
